```python
import cv2;
```

```python
import numpy as np;

```

```python
img = cv2.imread('app.jpg');
```

```python
from google.colab.patches import cv2_imshow ;
```

```python
cv2_imshow(img);
```

```python
w, h, c = img.shape
```

```python
img1 = img.copy()
```

```python
max_pix = np.max(img)
for i in range(w):
  for j in range(h):
    img1[i,j] = max_pix - img[i,j]
```

```python
cv2_imshow(img1)
```

```python
img_neg = ~img
```

```python
cv2_imshow(img_neg)
```
